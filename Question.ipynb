```python
#Alternate soln.
import numpy as np

a = np.arange(1,11,1)

indices = np.where((a>=3) & (a<=8))
    
for i in indices:
    a[i] = np.negative(a[indices])

for i in a:
    print(i)
```

Q18.)Terror Casualty Attack

```python
import numpy as np
import csv

with open("Dataset/terrorismData.csv",encoding = "ISO-8859-1") as file_obj:
    data = csv.DictReader(file_obj)

    info = []
    
    for row in data:
        if row['State'] == "Jammu and Kashmir" and row['Year'] == "1999" :
            info.append([row["City"],row["Month"],row["Killed"],row["Wounded"],row["Group"]])

    npInfo = np.array(info)
    
    npInfo = npInfo[np.logical_or(np.logical_or(npInfo[:,1] == "5" ,npInfo[:,1] == "6"),npInfo[:,1] == "7")]
    
#     npInfo[np.where(npInfo[:] == "")] = '0'
# or
# without using where.
    npInfo[npInfo == ""] = '0'
    
    killed = np.array(npInfo[:,2],dtype = float)
    wounded = np.array(npInfo[:,3],dtype = float) 
    cas = killed + wounded
    
    max_index = cas.argmax()
    
    print(int(cas[max_index]),npInfo[max_index,0],npInfo[max_index,4],end = " ")
```

Q20.)Terror City Casualty

```python
import numpy as np
import csv

with open("Dataset/terrorismData.csv",encoding = "ISO-8859-1") as file_obj:
    data = csv.DictReader(file_obj)

    info = []
    
    for row in data:
        if row['Country'] == "India":
            info.append([row["City"],row["Killed"],row["Wounded"]])

    npInfo = np.array(info)

    npInfo[npInfo == ""] = '0'
    
    killed = np.array(npInfo[:,1],dtype = float)
    wounded = np.array(npInfo[:,2],dtype = float) 
    cas = killed + wounded
    cas = cas.reshape(len(npInfo),1)

    npInfo = np.append(npInfo,cas,axis = 1)
    
    d = {}
    
    for i in npInfo:
        
        if i[0] == "Unknown":
            continue
            
        if i[0] in d:
            d[i[0]] += int(float(i[3]))
        else:
            d[i[0]] = int(float(i[3]))
            
    l = sorted(d,key = lambda item:d[item],reverse = True)
    
    for i in l[:5]:
        print(i,d[i],end = " ")
        print()
```

```python
import numpy as np
import csv

with open("Dataset/terrorismData.csv",encoding = "ISO-8859-1") as file_obj:
    data = csv.DictReader(file_obj)

    info = []
    
    for row in data:
            info.append(row["Day"])

    npInfo = np.array(info)

    npInfo[npInfo == ""] = '0'
    
    
    d = {}
    
    for i in npInfo:
        if i in d:
            d[i]+=1
        else:
            d[i] = 1
    
    m = max(d, key=d.get) 
    
    print(m,d[m],end = " ")
```
